```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
bat_data = pd.read_csv('/Users/gakuueno/Documents/Python/Python Data Analysis Projects/MLB Up to 2015/Batting.csv')

plate_appear = bat_data['AB'] + bat_data['BB'] + bat_data['HBP'] + bat_data['SF'] + bat_data['SH']
bat_data.insert(7, 'PA', plate_appear)

bat_avg = round(bat_data['H']/bat_data['AB'], 3)
bat_data.insert(18, 'BA', bat_avg)

obp = bat_data[['H', 'BB', 'HBP']].sum(axis=1)/bat_data[['AB', 'BB', 'HBP', 'SF']].sum(axis=1)
obp_rnd = round(obp, 3)
bat_data.insert(19, 'OBP', obp_rnd)

singles = bat_data['H'] - bat_data[['2B', '3B', 'HR']].sum(axis = 1)
bat_data.insert(10, '1B', singles)

slg = (bat_data['1B'] + 2*bat_data['2B'] + 3*bat_data['3B'] + 4*bat_data['HR'])/bat_data['AB']
slg_rnd = round(slg, 3)
bat_data.insert(20, 'SLG', slg_rnd)

ops = slg + obp
ops_rnd = round(ops, 3)
bat_data.insert(21, 'OPS', ops_rnd)

bat_data[['yearID','playerID', 'SF','H', '2B','OBP', 'SLG', 'OPS']]
```

```python
pitch_data = pd.read_csv('/Users/gakuueno/Documents/Python/Python Data Analysis Projects/MLB Up to 2015/Pitching.csv')
pitch_data.rename(columns = {'IPouts':'outs'}, inplace = True)
pitch_data.insert(13, 'IP', pitch_data['outs']/3)
pitch_data[['playerID', 'outs', 'IP', 'ERA']]
pitch_data.columns
```

```python
plt.figure(figsize=(13, 6))
sns.lineplot(x='yearID', y='SLG', data = bat_data)
```

```python
plt.figure(figsize=(13, 6))
sns.lineplot(x='yearID', y='ERA', data = pitch_data[pitch_data['IP'] > 150])
```
