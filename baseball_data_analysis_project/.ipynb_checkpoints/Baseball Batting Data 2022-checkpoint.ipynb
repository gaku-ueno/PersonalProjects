```python
#Goal: to find the correlation between different batting metrics and WAR, to see which metrics correlate with the most success

#Step 1: Collect data, get bref data for the 2022 season and create a database with it

#Step 2: Create plots of graphs between different player's WAR and batting statistics
```

```python
import pandas as pd
import csv
import re
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

Player Batting Statistics

```python
#Reading the batting data
batting_2022_frame = pd.read_csv('/Users/gakuueno/Documents/Python/Python Data Analysis Projects/MLB_2022_Batting.csv')
batting_2022_frame
```

```python
#Cleaning the data
batting_2022_frame = batting_2022_frame.drop('Name-additional', axis = 'columns').drop([992], axis = 'rows')
batting_2022_frame = batting_2022_frame.drop_duplicates(subset = 'Name')
batting_2022_frame['Name'] = batting_2022_frame['Name'].str.replace('[#*]', '') #removes the # and * that are next to the names
batting_2022_frame = batting_2022_frame.set_index('Name') #set name to index after you edit the names, because you can't change the index with the method you are using for some reason

print(len(batting_2022_frame))
batting_2022_frame
```

```python
#Selecting only qualified batters
batting_2022_frame = batting_2022_frame[batting_2022_frame['PA'] > 502]
print(len(batting_2022_frame))

batting_2022_frame
```

```python
batting_2022_frame.columns.tolist() 
```

```python
#Removing data
bat_data = batting_2022_frame.loc[:,['Age', 'G','PA','BA', 'OBP', 'SLG', 'OPS','OPS+','Pos Summary']]
bat_data
```

Player Value Statistics

```python
bref_value_data = pd.read_csv('/Users/gakuueno/Documents/Python/Python Data Analysis Projects/MLB_2022_Player_Value.csv')
bref_value_data = bref_value_data.drop('Name-additional', axis = 'columns')
bref_value_data['Name'] = bref_value_data['Name'].str.replace('[#*]', '')
bref_value_data = bref_value_data.set_index('Name')

bref_value_data.head(30)
```

```python
bref_value_data = bref_value_data[bref_value_data['PA'] > 502]
bref_value_data = bref_value_data.loc[:, ['WAR','WAA', 'oWAR', 'dWAR']]
bref_value_data
```

```python
player_data = pd.merge(bat_data,bref_value_data, on='Name')
player_data

#Converting the numbers and letters to easily readable positions
def position_converter(pos_sum):
    mapping = {
        '1': 'Pitcher',
        '2': 'Catcher',
        '3': 'First Base',
        '4': 'Second Base',
        '5': 'Third Base',
        '6': 'Shortstop',
        '7': 'Left Field',
        '8': 'Center Field',
        '9': 'Right Field',
        'D': 'Designated Hitter'
    }
    pos_eng = []

    for letter in pos_sum:
        if letter in ['*', '/']:
            continue
        position = mapping.get(letter, '')
        pos_eng.append(position)

    return '/'.join(pos_eng)

player_data['Pos Summary']= player_data['Pos Summary'].apply(lambda pos_sum: position_converter(pos_sum))
player_data['Pos Summary'] = player_data['Pos Summary'].apply(lambda x: re.sub(r'/+', '/', x))
player_data['Pos Summary'] = player_data['Pos Summary'].apply(lambda x: x.rstrip('/'))
player_data['Main Position'] = player_data['Pos Summary'].str.split('/').str[0]

player_data.head(10)
```

Explanation of the lambda functions:

```python
columns = player_data.columns.tolist()
columns.insert(9, columns.pop(columns.index('Main Position')))
player_data = player_data[columns]

player_data.head(20)
```

`.insert` is a method to add a element at a specific index, which is why we created a list in the line above <br>
`.pop` is a method that returns the element that was removed from the list <br>

```python
writer = pd.ExcelWriter('MLB_2022_Offense.xlsx', engine='xlsxwriter')

# Write the DataFrame to the Excel file
player_data.to_excel(writer, sheet_name='Sheet1')

# Save the Excel file
writer.save()
```

```python
sns.regplot(x='OPS+', y='WAR', data=player_data, scatter=True)
slope, intercept, r_value, p_value, std_err = stats.linregress(player_data['OPS+'], player_data['WAR'])
print("Slope:", slope)
```
