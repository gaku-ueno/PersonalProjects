```python
import pandas as pd
import numpy as np

#importing batting data first
from pybaseball import batting_stats, batting_stats_bref, bwar_bat, statcast_batter, playerid_lookup, statcast_batter_expected_stats, statcast_batter_exitvelo_barrels
from pybaseball import cache

cache.enable()

from datetime import date

from bokeh.plotting import figure, output_notebook, show
from bokeh.models import ColumnDataSource, HoverTool
```

```python
def scatter_linregress_plot(x_column, y_column, data_source):
    x = x_column.astype(float)
    y = y_column.astype(float)
    
    source = ColumnDataSource(data_source)
    
    tooltips=[
        (f'{x_column.name}', '@{'f'{x_column.name}''}'),
        (f'{y_column.name}', '@{'f'{y_column.name}''}'),
    ]
    
    yr_value_plot = figure(title = f'Relationship between {x_column.name} and {y_column.name}', x_axis_label = f'{x_column.name}', y_axis_label = f'{y_column.name}', tooltips = tooltips)
    yr_value_plot.circle(f'{x_column.name}', f'{y_column.name}', size = 5, source = data_source)

    coefficients = np.polyfit(x, y, 1)
    slope = coefficients[0]
    intercept = coefficients[1]
    y_values = slope * x + intercept
    
    yr_value_plot.line(x, y_values, line_color = 'red')

    output_notebook()
    show(yr_value_plot)
    print(f'Slope: {slope}')
```

## Plotting how a player's performance changes over time
- Plot how performance changes with age
- Plot how a single player's, say Albert Pujols, performance changes with age

```python
#Plotting how performance changes with age
bat_stat = batting_stats(2019, end_season = 2022, qual = 400)
pujols = bat_stat[bat_stat['Name'] == 'Albert Pujols']
pujols
```

```python
bat_stat = batting_stats(1990, end_season = 2022, qual = 200)
bat_stat = bat_stat[['Name', 'Age', 'WAR']]

bat_stat = bat_stat[(bat_stat['Age'] < 40) & (bat_stat['Age'] > 20)]

name = bat_stat['Name']
fwar = bat_stat['WAR']
age = bat_stat['Age']

poly_function = np.poly1d(np.polyfit(age, fwar, 2))
polyline = np.linspace(min(age), max(age), 100)

plot = figure(title = 'Age vs WAR', x_axis_label = 'Age', y_axis_label = 'fWAR')
plot.circle(age, fwar, size = 5)
plot.line(polyline, poly_function(polyline), line_width=2, color='red', legend_label='Quadratic Line') #the polyline values are passed into the poly_function which creates a smooth graph

output_notebook()
show(plot)

print(poly_function)
```
